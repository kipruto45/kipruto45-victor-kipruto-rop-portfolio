# Equity Group: Mobile Banking & Product Analytics
Analysis of Equitel subscriber growth, EazzyPay volumes, and product cross-sell performance.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

engine = create_engine('postgresql://equity_admin:equity_password@localhost:5435/equitel_analytics')

adoption = pd.read_sql('SELECT * FROM mart_adoption_curve ORDER BY period', engine)
arpu = pd.read_sql('SELECT * FROM mart_arpu_benchmark ORDER BY period', engine)
cross_sell = pd.read_sql('SELECT * FROM mart_cross_sell_rate ORDER BY period', engine)

print("Loaded Equitel analytical marts.")
```

## 1. Subscriber Growth & EazzyPay Adoption
Tracking the growth of transaction velocity post-adoption.

```python
plt.figure(figsize=(12, 5))
plt.plot(adoption['period'], adoption['transaction_count'], marker='o', color='green', label='Transaction Count')
plt.title('EazzyPay Adoption Curve')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## 2. Product Cross-Sell Rates
Analyzing the penetration of Insurance and Investment products within the Equitel base.

```python
plt.figure(figsize=(12, 5))
plt.plot(cross_sell['period'], cross_sell['insurance_cross_sell_rate'], label='Insurance %')
plt.plot(cross_sell['period'], cross_sell['investment_cross_sell_rate'], label='Investments %')
plt.title('Cross-Sell Rate: Insurance & Investments')
plt.ylabel('Penetration %')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

## 3. ARPU Benchmark
Monthly Average Revenue Per User trend.

```python
plt.figure(figsize=(12, 5))
plt.bar(arpu['period'], arpu['arpu_kes'], color='orange')
plt.title('Equitel: Monthly ARPU Trend (KES)')
plt.xticks(rotation=45)
plt.show()
```
